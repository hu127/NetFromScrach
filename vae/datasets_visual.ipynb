```python
import torch
import torchvision.datasets as datasets
```

```python
training_data = datasets.CIFAR10(root="data", train=True, download=True)
```

```python
# visualize the data
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 5)
for i in range(5):
    ax[i].imshow(training_data[i][0])
    ax[i].set_title(training_data[i][1])
    ax[i].axis("off")
plt.show()
```

```python
# visualize the data format
print("data size: ", len(training_data))
print("data format: ", training_data[0])
print("data type: ", type(training_data[0][0]))
print("data label: ", training_data[0][1])
print("data label type: ", type(training_data[0][1]))
```

```python
# make data normalize
from torchvision.transforms import transforms

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (1.0, 1.0, 1.0)) # mean, std
])

training_data = datasets.CIFAR10(root="data", train=True, download=True, transform=transform)
```

```python
# visualize the data
fig, ax = plt.subplots(1, 5)
for i in range(5):
    ax[i].imshow(training_data[i][0].permute(1, 2, 0))
    ax[i].set_title(training_data[i][1])
    ax[i].axis("off")
plt.show()
```
